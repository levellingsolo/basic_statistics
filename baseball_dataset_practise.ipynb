```python
#libraries
import pandas as pd
import seaborn as sns
```

```python
#load the dataset
df = pd.read_csv("Baseball.csv")
df.head()
```

#mean

```python
#Mean of Run Scored(RS)
df.RS.mean()
```

```python
#mean of run scored (RS) by arizona team
df.loc[df.Team == "ARI"].RS.mean()
```

```python
#Mean of Runs Scored (RS) by the arizona team (ARI) since 2005
df.loc[(df.Team == "ARI") & (df.Year > 2005)].RS.mean()
```

```python
#what is the mean of runs allowed (RA) by chicago team (CHC) till 2007
df.loc[(df.Team == "CHC") & (df.Year < 2007)].RA.mean()
```

#Median and Mean of Wins(W)


```python
print(df.W.mean())
df.W.median()
```

```python
#Question -  what is the median of wins of the Baltimore team (BAL)  untit 2000
df.loc[(df.Team == "BAL") & (df.Year <= 2000)].W.median()
```

```python
#mode , median and mean of OBP
print(df.OBP.mean())
print(df.OBP.median()
print(df.OBP.mode())
```

```python
df.loc[df.Year == 2010]
```

```python
#mode of OBP during the year 2010
df.loc[df.Year == 2010].OBP.mode()
```

#Correlation

```python
#pick variables
df_correlation = df[["RS", "RA", "W"]]
df_correlation.head(1)
```

```python
#Correlation_Matrix
df_correlation.corr()
```

```python
#Correlation heatmap
sns.heatmap(data = df_correlation.corr(),
            annot = True,                      #  annot is for values in the grapn
            fmt = '.2g',
            center = 0 ,
            cmap = 'coolwarm',
            linewidths = 1,
            linecolor = 'black'
            )
```

```python
#challenge  :  correlation matrix between OBP , SLG and BA.   do as well a heatmap
#pick variables
df_correlation1 = df[["OBP", "SLG", "BA"]]
df_correlation1.head(2)
```

```python
#Correlation matrix
df_correlation1.corr()
```
